# Question 2: Naïve Bayes

Training, optimisation, and evaluation of machine learning systems should be conducted on different data sets.
<br>
The data will be split in two disjoint sets:

<strong>1 -</strong> a training set that will be use dto construct the classifier <br>
<strong>2 -</strong> a test set, hat will be used to test the classifier.

The 33% of the data for <strong>Testing</strong> and the rest for <strong>Training</strong>. <br>
The random number generator to a defined value so that we'll get the same split each time we run this code. <br>

Data is split after it has been encoded as x and y. <br>
This is to ensure that the test and training data are encoded in the same way.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,precision_score,accuracy_score,recall_score,classification_report, accuracy_score
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=1234)
print("X_Train: " ,X_train.shape, "X_Test: " ,X_test.shape)
print("Y_Train: " ,y_train.shape, "Y_Test: " ,y_test.shape)
```

BernoulliNB is one of many Scikit-learn classifiers and it is one of the simple Naive Bayes classifier <br>

call the fit(x,y) method to train the classifier

```python
from sklearn.naive_bayes import BernoulliNB
BernNB = BernoulliNB()
BernNB.fit(X_train,y_train)
print(BernNB)
```

As part of Evaluation, first predict the target labels from the feature matrix, and then compare the predicted target labels to the correct target labels.<br>

From that we can derive measures such as accuracy, and we can construct contingency tables to see where the classifier makes the most errors.



```python
y_pred = BernNB.predict(X_train)

print('Correctly predicted on set: {}'.format(sum(y_train == y_pred)))
print('errors: {}'.format(sum(y_train != y_pred)))
```

The accuracy,Precision and Recall result of tranning set are listed bellew:

```python
print(classification_report(y_train, y_pred))
print('Accuracy on TRAINING set: {:.4f}'.format(accuracy_score(y_train, y_pred)))
print('Precision on TRAINING set: {:.4f}'.format(precision_score(y_train,y_pred)))
print('Recall on TRAINING set: {:.4f}'.format(recall_score(y_train,y_pred)))
```

The purpose of confusion matrix is to display the details about the classifers.<br>
The columns in the matrix represent the predicted classes, and the rows represent the correct classes.<br>
classifier predicts 942 zeros and 292 ones. <br>

```python
from nltk import ConfusionMatrix
print("Confusion Matrix:\n")
print(ConfusionMatrix(list(y_train), list(y_pred)))
```

This shows how well the classifier works on the TEST set.

```python
y_pred = BernNB.predict(X_test)
print('Correctly predicted on TEST set: {}, errors: {}'.format(sum(y_test==y_pred), sum(y_test!=y_pred)))
print('Accuracy on TEST set: {:.2f}'.format(accuracy_score(y_test,y_pred)))
```

Using the method predict_proba lets see how probable each target label may be for the first test sample. 

```python
y_pred_proba = BernNB.predict_proba(X_test)
print(y_pred_proba[0])
```

```python
min_samples_leaf = np.arange(1,5)
max_depth =  np.arange(1,30)
max_leaf_nodes =  np.arange(2,20)
parameters = {'max_depth': max_depth,'min_samples_leaf': min_samples_leaf, 'max_leaf_nodes': max_leaf_nodes}
print(parameters)
```

```python
GridSearchCV = GridSearchCV(DecisionTreeClassifier(), parameters)
GridSearchCV.fit(X_train, y_train)
print(GridSearchCV)

maxDepth = GridSearchCV.best_params_["max_depth"]
minLeaf = GridSearchCV.best_params_["min_samples_leaf"]
maxLeaf = GridSearchCV.best_params_["max_leaf_nodes"]
```
